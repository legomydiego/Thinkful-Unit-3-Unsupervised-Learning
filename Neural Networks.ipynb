```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('economic_freedom_index2019_data.csv', encoding = 'ISO-8859-1')
```

```python
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()/len(df)*100
```

```python
df[df.isnull().any(axis=1)]
```

```python
df.dropna(inplace=True,axis=0)
```

```python
df.info()
```

```python
df['Population (Millions)'] = pd.to_numeric(df['Population (Millions)'])
```

```python
df['GDP (Billions, PPP)'] = pd.to_numeric(df['GDP (Billions, PPP)'].replace('[\$,]', '', regex=True).astype(float))
```

```python
df['GDP per Capita (PPP)'] = pd.to_numeric(df['GDP per Capita (PPP)'].replace('[\$,]', '', regex=True).astype(float))
```

```python
df['Unemployment (%)'] = pd.to_numeric(df['Unemployment (%)'])
```

```python
df['FDI Inflow (Millions)'] = pd.to_numeric(df['FDI Inflow (Millions)'].replace('[\$,]', '', regex=True).astype(float))
```

```python
df.info()
```

```python
df.drop(['CountryID','Country Name','WEBNAME'], axis=1, inplace=True)
```

```python
df.head()
```

```python
dfDummies = pd.get_dummies(df['Region'], prefix = 'Region')
```

```python
df = pd.concat([df, dfDummies], axis=1)
```

# Modeling

```python
X = df.drop(['Region','Country','World Rank','Region Rank','2019 Score'],axis=1)
y = df['World Rank']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score
from sklearn import metrics
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

Random Forest

```python
from sklearn.model_selection import GridSearchCV
from sklearn.ensemble import RandomForestRegressor
```

```python
gridsearch_forest = RandomForestRegressor()

params = {
    "n_estimators": [50,60,70],
    "max_depth": [18,19,20],
    "min_samples_leaf" : [1, 2, 3]
}

clf = GridSearchCV(gridsearch_forest, param_grid=params, cv=5 )
clf.fit(X_train,y_train)
```

```python
clf.best_params_
```

```python
clf.best_score_
```

```python
cross_val_score(clf, X, y, cv=10)
```

```python
from sklearn.ensemble import GradientBoostingRegressor
```

```python
clf = GradientBoostingRegressor(n_estimators=60,max_depth=20,min_samples_leaf=2)
clf.fit(X_train, y_train)
```

```python
print(clf.score(X_test, y_test))
```

```python
cross_val_score(clf, X, y, cv=10)
```

Neural Network

```python
from sklearn.neural_network import MLPRegressor
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()  
scaler.fit(X_train)
X_scale = scaler.transform(X)
X_train_scale = scaler.transform(X_train)  
X_test_scale = scaler.transform(X_test)
```

```python
mlp = MLPRegressor(hidden_layer_sizes=(10000),random_state=101, learning_rate='adaptive', max_iter=1000)
```

```python
mlp.fit(X_train_scale,y_train)
```

```python
print(mlp.score(X_test_scale,y_test))
```

```python
cross_val_score(mlp, X_scale, y, cv=10)
```
